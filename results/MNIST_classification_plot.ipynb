```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress, spearmanr
from matplotlib import rcParams
```

```python
df = pd.read_csv("MI_vs_linear_probe.csv")
```

```python
fig = plt.figure(figsize=(3,2), dpi=200)

sns.set_style('whitegrid')
# scatterplot!
sns.lineplot(data=df, x='linear', y='MI', marker='o')

spearman = spearmanr(df['linear'], df['MI'])[0]


# show r on plot
plt.text(df['linear'].min(), df['MI'].max(), f"Spearman = {spearman:.2f}", ha='left', va='top', fontsize=10)

plt.ylabel('Mutual information')
plt.xlabel('Linear classification accuracy')
plt.show()
```
