```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

```python
df = pd.read_csv('PBMC_CITEseq_Harmony_scaling.csv')

old_df = pd.read_csv('PBMC_CITEseq_scaling.csv')
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(3, 2), dpi=200)

sns.lineplot(data=df, x='UMI per cell', y='Protein MI', marker='o', label='Harmony')
sns.lineplot(old_df[(old_df['Fraction'] == 1.) & (old_df['Method'] == 'VAE')],
             x='UMI per cell', y='Protein MI', marker='o', color='red', label='Uncorrected')
plt.xscale('log')
```
