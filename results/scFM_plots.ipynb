```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
sns.set_style('whitegrid')
rcParams['figure.dpi'] = 150
rcParams['figure.figsize'] = (4, 3)
rcParams['grid.linewidth'] = 0.5
rcParams['grid.linestyle'] = 'dashed'
rcParams['legend.fancybox'] = False
rcParams['mathtext.fontset'] = 'stix'

dpi = 150

hue_order = ["VAE", "PCA", "nanoTxformer", "RandomProjection"]
c6 = ['#c4bae2','#40ada6','#3c4ebf',  
        '#e3a600', '#edb1a3', '#d6543a',]
pretty_palette = [c6[c] for c in range(6) if c in [1,2,3,5]]
```

```python
df = pd.read_csv("../results/PBMC_CITEseq_inference_time_scaling.csv", index_col=0)
df2 = pd.read_csv("../results/PBMC_CITEseq_inference_time_scaling_nanoTx.csv", index_col=0)
df = pd.concat([df, df2])

df.replace('Random projection', 'Rand. proj. (128 dim)', inplace=True)
df.replace('Rank random projection', 'Rank rand. proj. (128 dim)', inplace=True)
```

```python
ax = sns.lineplot(data=df, x='Quality', y='MI', hue='Model',
             marker='o', palette=c6[1:])
sns.move_legend(ax, loc=(1.05, 0.45))

plt.xscale('log')
```
